```python
# 剔除客户编号为空的数据
data_.dropna(subset=['客户编号'], inplace=True)
```

```python

# 剔除重复数据
data_.drop_duplicates(keep="first",inplace=True)
# 取出单位售价大于0的商品
data_1=data_[data_['单位价格（英镑）']>0]
#查看数据基本情况
data_1.info()
```

```python
#生成购物总金额列
data_1['总金额']=data_['销售数量']*data_['单位价格（英镑）']

# 客户编号不需要用来计量顾转换类型
data_1["客户编号"]=data_1["客户编号"].astype('object')

# 为了剔除非销售所获收入定义函数
def action(x):
    try:
        if type(x)==int:
            return x
        elif type(int(x[:-1]))==int:
            return x
    except:
        return np.nan
data_1["新产品代码"]=data_1["产品代码"].apply(action)

# 剔除非产品销售收入的行
data_1=data_1[data_1["新产品代码"].notnull()]
data_1
```

```python
data_1["交易日期"].max()
```

```python
data_1
```

```python
# 提取时间列
data_1["交易日期"].max()
data_1["观察日期"]=pd.to_datetime('2011-12-11 00:00:00')
# 求得最后一次发生购物行为距观测窗口结束的时长
delta=(data_1["观察日期"]-data_1["交易日期"])
data_1['datedif_']=delta.apply(lambda x : x/np.timedelta64(30,'D'))
```

```python
data_2=data_1[['客户编号','销售数量','总金额','datedif_']]
data_2["freqency"]=1
data_2.head()
```

```python
# 将数据表中的数据进行聚合
data_3=data_2.groupby("客户编号").aggregate({"销售数量":np.sum,"总金额":np.sum,"datedif_":np.min,"freqency":np.sum})
# 剔除销售金额与销售数量为负的记录
data_3=data_3[(data_3["销售数量"]>0)&(data_3["总金额"]>0)]
data_3["购买数量/次"]=data_3["销售数量"]/data_3["freqency"]
data_3.shape
```

```python
# 提取RFM模型所需字段
data_3=data_3.iloc[:,1:4]
data_3.head()
```

```python
data_4=data_3
```

构造RFM指标：

- R=datedif_(最后一次发生购物行为距观测窗口结束的时长)
- F=购买数量/次(观测窗口内的每次购物购买的平均件数)
- M=total_money(观测窗口内的订单总金额)

```python
#计算每个字段的均值与标准差  
mean_t=np.array(data_3.mean())
std_t=np.array(data_3.std())

# Z-Score标准化
Z_data_3=(data_3-mean_t)/std_t
Z_data_3.head()
```

##  建立模型

### 主要是两个部分：
* 利用K-means算法对客户进行聚类。
* 结合业务，分析客户特征，分析客户价值。

- KMeans

```python
#### 导入K均值聚类算法
from sklearn.cluster import KMeans
def distance_(v1, v2):                  
    return np.sum(np.power(v1 - v2, 2))                                       #计算欧氏距离
def test_Kmeans_nclusters(data_train):                                         #计算最适合的簇
    data_train = data_train.values                                             #将数据从DataFrame中提取数据np.array
    num = range(2, 10)
    SSE = []
    for i in num:                                                           #建立循环计算分成2个簇-9个簇的sse大小
        sse = 0
        kmodel = KMeans(n_clusters=i)                                        #建立KMeans模型
        kmodel.fit(data_train)                                                  #训练数据
        
        cluster_ceter_list = kmodel.cluster_centers_                            #求得质心
        cluster_list = kmodel.labels_.tolist()                                #求得每个样本属于哪一个簇
        for index in range(len(data_3)):
            cluster_num = cluster_list[index]
            sse += distance_(data_train[index, :], cluster_ceter_list[cluster_num])
        print("簇数是", i, "时； SSE是", sse)
        SSE.append(sse)
    return num, SSE
num, SSE = test_Kmeans_nclusters(Z_data_3)
```

```python
## 绘图观测SSE与簇个数的关系
plt.figure(figsize=(10,8))
plt.plot(num,SSE,marker='o',ls=':')
plt.xlabel("n_clusters", fontsize=18)
plt.ylabel("SSE", fontsize=18)
plt.title("KMeans", fontsize=20)
plt.show()
```
