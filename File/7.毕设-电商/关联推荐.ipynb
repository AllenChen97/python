```python
#剔除发票编号以字母“C”开头的数据
data_01=data_01[data_01.发票号码.apply(lambda x:str(x)[0]!="C")]
```

```python
#剔除销售数量为负的
data_01=data_01[data_01.销售数量>0]
```

```python
#数据重排，生成事务数据集
basket = data_01.groupby(['发票号码','品类名称'])['销售数量'].sum().unstack()
basket
```

```python
#不关心产品销售数量，把大于零的替换为1，否则为零
basket = basket.applymap(lambda x: 1 if x>0 else 0)
```

```python
basket.head(1)
```

```python
import seaborn as sns

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#设置最小支持度为0.03，支持度越大越好，所有规则最大支持度仅为0.32
import datetime
a=datetime.datetime.now()
itemsets = apriori(basket, min_support=0.015, use_colnames=True)
datetime.datetime.now()-a
```

```python
itemsets.shape
```

```python
itemsets.sort_values('support',ascending=False).head()
```

```python
#生成关联规则，最小提升度设置为1，提升度大于1才认为该规则有效
rules = association_rules(itemsets, metric="lift", min_threshold=1)
```

```python
rules.shape
```

```python
#散点图
plt.figure(figsize=(9,5.5))
sns.scatterplot(x='support', y='confidence', hue='lift', data=rules)
plt.tick_params(labelsize=17)
plt.xlabel("support",size=17)
plt.ylabel("confidence",size=17)
plt.show()
```

```python
#由散点图，进行筛选运算
selected_rules = rules[ (rules['lift'] >= 4) & (rules['confidence'] >= 0.5) ]
print('selected rule num is:', len(selected_rules))
selected_rules.sort_values(by='lift',ascending=False)
```

```python
plt.figure(figsize=(9,5.5))
sns.scatterplot(x='support', y='confidence', hue='lift', data=selected_rules)
plt.tick_params(labelsize=17)
plt.xlabel("support",size=17)
plt.ylabel("confidence",size=17)
plt.show()
```

```python
selected_rules=selected_rules.sort_values(by='lift',ascending=False)
```

```python
def func(x):
    a=[]
    for i in x:
        b=[]
        for j in i:
            b.append(j)
            c=','.join(b)
            c= c.strip(",")
        a.append(c)
    return a
```

```python
selected_rules['antecedents']=func(selected_rules.antecedents)
selected_rules['consequents']=func(selected_rules.consequents)
```
