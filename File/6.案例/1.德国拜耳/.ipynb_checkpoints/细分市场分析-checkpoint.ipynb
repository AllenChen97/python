```python
dataset=[dwxtz,fmfzp,ssscj,pxmzxwxp,wxjrq,wxp,wxy]
def timeindex(x):
    x.index=(x.时间)
    del x["时间"]
for i in dataset:
    timeindex(i)
```

```python
sales_figure3 = pd.concat([dwxtz,fmfzp,ssscj,pxmzxwxp,wxjrq,wxp,wxy],axis=1)
sales_figure3.columns = ["电蚊香套装","防霉防蛀片","灭鼠杀虫剂","盘香","蚊香加热器","蚊香片","蚊香液"]
sales_figure3 = (sales_figure3/10000).round(2)
```

```python
sales_figure3.head(5)
```

```python
def forecast(n):#补全2018-11/12数据
    
    from sklearn import linear_model
    import datetime
    a=sales_figure3["2015-%d-01"%(n)]
    b=sales_figure3["2016-%d-01"%(n)]
    c=sales_figure3["2017-%d-01"%(n)]
    d= pd.concat([a,b,c])

    empty=[]
    x_train = np.array([2015,2016,2017]).reshape(-1,1)
    linear_reg = linear_model.LinearRegression()
    for i in range(7):
        y_train = np.array(d.iloc[:,i])
        linear_reg.fit(x_train,y_train)
        y_predict = linear_reg.predict(np.array([2018]).reshape(-1,1))
        empty.append(y_predict[0])
        
    time=datetime.datetime.strptime("2018-%d-01"%(n),"%Y-%m-%d")
    sales_figure3.loc[time]=empty
    
forecast(11)
forecast(12)

sales_figure3 = sales_figure3.sort_index(ascending=False).round(2)#排序
sales_figure4 = sales_figure3.iloc[:-2,:]#不看2015-11/12
sales_figure4.tail(5)
```

### 年销量

```python
import datetime
sales_year = sales_figure4.resample(rule='Y').sum()#重采样
sales_year
```

### 增速

```python
def growth(x):
    empty_big = []
    for i in range(7):
        empty = []
        for j in range(3):
            b=(x.iloc[j,i]/x.iloc[0,i]-1)*100
            empty.append(b.round(2))
        empty_big.append(empty)
    return empty_big

sales_growth = pd.DataFrame(growth(sales_year),index=sales_year.columns,columns=sales_year.index).T
sales_growth
```

```python
# 解决坐标轴刻度负号乱码
plt.rcParams['axes.unicode_minus'] = False
# 解决中文乱码问题
plt.rcParams['font.sans-serif'] = ['Simhei']
sales_growth.plot(kind='line',figsize=(8,4),fontsize=12)
```

```python
a=(np.sqrt(sales_growth.iloc[2,:]/100+1)-1)*100 #年复合增长
a

b=sales_year.apply(lambda x:np.mean(x))/10000 #平均销量(亿)
b

b=pd.DataFrame(b,columns=['年平均销量（亿）'])#合并
b["年复合增长（%）"]=a 
b.round(2)
```

### 季节性波动

```python
import datetime
sales_season = sales_figure4.resample(rule='3MS').sum()#重采样
sales_season=sales_season.drop(index=datetime.datetime.strptime("2019-01-01",'%Y-%m-%d'))
sales_season
```

```python
sales_season.plot(kind='line',figsize=(8,4),fontsize=12)
```

```python
def percentage(x): #市场体量占比
    empty_big = []
    for i in range(7):     
        empty = []
        for j in range(len(x)):

            save = x.iloc[j,i]/(0.01*list(x.T.apply(lambda b:sum(b)).round(2))[j]) #当季卖的量/当季所有一级市场总销量
            empty.append(save.round(2))
        empty_big.append(empty)
    return empty_big
```

```python
sales_percentage = pd.DataFrame(percentage(sales_season),index=sales_season.columns,columns=sales_season.index).T
sales_percentage.plot(kind='line',figsize=(8,4),fontsize=12)
```

```python
sales_percentage_y = pd.DataFrame(percentage(sales_year),index=sales_year.columns,columns=sales_year.index).T
sales_percentage_y.plot(kind='line',figsize=(8,4),fontsize=12)
```

1.在一级市场的季节表现来看，有明显的季节性波动，猜测产品需求和季节天气有较高的相关性。

2.经检查，在3年的数据里，各级别市场在整体灭虫市场中的占比并没有太大的变化。

3.有趣的是，从3年的占比数据中都能验证第一大市场（灭鼠杀虫剂市场）和第二大市场（蚊香液市场）在季节性波动中呈相反的轨迹，建议在资源投放的过程中可以根据波动规律，做细微的市场战略调整：Q1-Q2资源投放可以逐渐从第一市场倾斜至第二市场；Q3-Q4资源投放又可以逐渐从第二市场转回第一市场。但主策略还是需要高度关注第一大市场的资源投放。

### 价格分析

```python
def groupby(x,cata):
    x = x[["类别","销量（人数）","预估销售额","售价","收藏人数","评价人数"]][data.类别==cata]#取数
    x["价格段"]=pd.qcut(x.售价,4) #标记价格段


    x = x.groupby("价格段").sum() #聚合
    x["宝贝数"] = x.groupby("价格段").count().iloc[:,0] #添加宝贝数

    x=x.drop(columns="售价")#去除售价（没用）
    return x
```

```python
def calculating(x):
    #计算
    x["销量（人数）"] = x["销量（人数）"]/x["宝贝数"]

    #数值处理
    x['预估销售额'] = x['预估销售额']/10000
    x['收藏人数'] = x['收藏人数']/10000
    x['评价人数'] = x['评价人数']/10000

    #列命名、排序整理、去除宝贝数（已经没用了）
    x.columns = ["宝贝平均客户数",'预估销售量（万）','收藏热度（万人）','评价热度（万人）','宝贝数']
    x = x.round(2).sort_values(by='预估销售量（万）',ascending=False)
    
    return x[['预估销售量（万）',"宝贝平均客户数",'收藏热度（万人）','评价热度（万人）']]
```

```python
list=['灭鼠','蟑螂','虱']

for i in list:
    calculating(groupby(data,i))
```

1.灭鼠市场的第一大价格区间在(52.6,498]，但平均宝贝客户数明显要比第二价格区间(0,15.8]低，所以灭鼠产品在(0,15.8]区间内，竞争会更为缓和，市场进入的容易程度更高。

2.灭蟑市场的第一大价格区间在(34.4,500]，其竞争激烈程度与第三价格区间(20,34.3]相近，市场进入容易程度最高的是第二价格区间(12.8,20]，所以进入/布局灭蟑市场优先在第二价格区间(12.8,20]。

3.除虱市场的第一大价格区间(23.9,37.8]的平均宝贝客户数将其他价格区间以2.5倍以上的差距远抛其后，优先选择(23.9,37.8]用于产品定价。

物理形态仅存在拜耳的数据集中，而在数据集中并未显现适用对象间的差异性，所以放弃该字段。

```python
def five_g(x): #净含量字段整理，用于下面applymap
    if '+5ml*4'in x:
        x=x.replace('+5ml*4','')
    if '+12g'in x:
        x=x.replace('+12g','')
    return x
```

```python
#统一净含量格式（去除赠品）：提取-转格式-applymap修改
sales30_be3 = sales30_be3[sales30_be.适用对象=='蟑螂'].astype('str').applymap(five_g)
#提取净含量和销售量,并转换为整数型
sales30_be3 = sales30_be3.iloc[:,2:]
sales30_be3['30天销售量']=sales30_be3['30天销售量'].astype("int")
sales30_be3['总销量']=sales30_be3['总销量'].astype("int")
#聚合、排序
sales30_be3.groupby('净含量').sum().sort_values('30天销售量',ascending=False)
```

## 评论数据文本挖掘

```python
context_be = pd.read_excel("拜耳评论数据.xlsx")
context_as = pd.read_excel("安速评论数据.xlsx")
context_kl = pd.read_excel("科凌虫控评论数据.xlsx")
```

```python
def data_p(x):
    x.index=x["评论日期"]
    x = x[["评论"]]
    x.columns=["comments"]
    return x
context_be = data_p(context_be)
context_as = data_p(context_as)
context_kl = data_p(context_kl)
```

```python
def cleaning(x):     #处理不要的评论
    b=[]
    for i in x:           
        while r"&hellip;"in i or r"hellip" in i:
            i = i.replace(r"&hellip;","")
        while '哎哎'in i:
            i = i.replace("哎哎","")
        if (i =="此用户没有填写评论!") or (i in "好评好评好评") or ("还没用"in i):
            continue
        b.append(i)
    return b
```

```python
def process_write(x,name):
    x = cleaning(x["comments"]) #掉入Series 出list
    
    for comment in x:     #下载文本
        with open('%s_comments.txt' % name,'a',encoding='utf-8') as f:
            f.write(comment + '\n')
    return x
```

```python
lis=[context_be,context_as,context_kl]
list_name=['be','as','kl']
for data,name in zip(lis,list_name):
    lis[list_name.index(name)]=process_write(data,name)
lis
```

### 情感分析

```python
def stm(x):
    import numpy
    from snownlp import SnowNLP
    
    score_empty=[]
    for i in x:
        stms = SnowNLP(i).sentiments
        score_empty.append(stms) #将每句话的评分储存在列表
    return np.mean(score_empty).round(2)
```

```python
S={}
for data,name in zip(lis,list_name):
    S[name] = stm(data)
S
```

从评论的情感得分来看，拜耳产品的评论略消极于其对手安速和科凌，说明用户对于产品或服务更不满意，其原因还需要从评论中继续挖掘。
