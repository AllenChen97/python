```python
#完整代码，爬取一页
import requests
import pandas as pd 
from bs4 import BeautifulSoup
url='https://movie.douban.com/subject/26322642/comments?start=0&limit=20&sort=new_score&status=P'
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'}
response = requests.get(url,headers=headers)
response.encoding='utf-8'
soup = BeautifulSoup(response.text,'lxml')#对html进行解析，完成初始化
itemslist=[]
html=soup.select(' .comment-item')
for item in html:
    it=[]
    it.append(item.find('div',class_="avatar").a['title']) #评论人    
    it.append(item.find('span',class_="comment-time")['title']) #评论时间
    it.append(item.find('span',class_="short").text) #评论内容
    it.append(item.find('span',class_="votes").text) #点赞数
    itemslist.append(it)
pd.DataFrame(itemslist,columns=['评论人','评论时间','评论内容','点赞数']).set_index('评论人')
```

# 爬取多页

```python
#完整代码，爬取多页
import requests
import pandas as pd 
from bs4 import BeautifulSoup
import time

data=pd.DataFrame()
for i in range(10):
    print('正在爬取第%d页'%i)
    url='https://movie.douban.com/subject/26322642/comments?start='+str(i*20)+'&limit=20&sort=new_score&status=P'
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'}
    response = requests.get(url,headers=headers)
    response.encoding='utf-8'
    soup = BeautifulSoup(response.text,'lxml')#对html进行解析，完成初始化
#解析请求结果得到网页源码
    
    itemslist=[]
    html=soup.select(' .comment-item')
    for item in html:
        it=[]
        it.append(item.find('div',class_="avatar").a['title']) #评论人    
        it.append(item.find('span',class_="comment-time")['title']) #评论时间
        it.append(item.find('span',class_="short").text) #评论内容
        it.append(item.find('span',class_="votes").text) #点赞数
        itemslist.append(it)
        df=pd.DataFrame(itemslist)
    data=pd.concat([data,df])
    time.sleep(6)
#把
    data.columns=['评论人','评论时间','评论内容','点赞数']
data.set_index('评论人')
```

# 下载多个页面的代码

```python
from urllib.parse import urlencode
from hashlib import md5
import requests
import time
import os
os.chdir(r'C:\Users\CDA')
import pandas as pd 
from bs4 import BeautifulSoup

def get_page(offset):#设置requests的参数
    params = {
        'aid': '24',
'app_name': 'web_search',
'offset': offset,
'format': 'json',
'keyword': '街拍',
'autoload': 'true',
'count': '20',
'en_qc': '1',
'cur_tab': '1',
'from': 'search_tab',
'pd': 'synthesis'    }
    base_url = 'https://www.toutiao.com/api/search/content/?aid=24&app_name=web_search&offset='
    base_url2 ='0&format=json&keyword=%E8%A1%97%E6%8B%8D&autoload=true&count=20&en_qc=1&cur_tab=1&from=search_tab&pd=synthesis&timestamp=1568085504647'
    url = base_url + urlencode(params) + base_url2
    return url

for i in range(5):  #不同页面，多页
    url=get_page(i*20)
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36','cookie':'tt_webid=6699246252980225544; UM_distinctid=16b2ab806ac249-0136f7c490c699-3e385b04-100200-16b2ab806ad293; csrftoken=ca1c6ada367837859b79501b56312c1a; __tasessionId=39azmsspu1559789837584; CNZZDATA1259612802=1112984953-1559786375-https%253A%252F%252Flanding.toutiao.com%252F%7C1559786375; s_v_web_id=eaecbccae6651f52de0c381ee9e32670'}
    response = requests.get(url,headers=headers)
    html=response.json()
    data=html.get('data')
#解析后，获取多个页面的的data节点

    for item in data:    
        if item.get('cell_type') is not None:#跳过data下的0节点
            continue
        urllist=item['image_list']
        title=item['title']
        title=validatetitle(title)
        print(title)
#得到所有的image_url 和 处理后的title名字
    
        for url in urllist:#一个title下不同的图片
            newurl=url['url'].replace('list/190x124','large')
            print(newurl)
            resp = requests.get(newurl)#get图片地址？ 为了写哈希编码作为图片名字
            file_path = 'img' + os.path.sep + title
            if not os.path.exists(file_path):#没有该文件夹就创一个
                os.makedirs(file_path)
            
            img_path = file_path + os.path.sep + '{}.jpg'.format(md5(resp.content).hexdigest())
            
            with open(img_path, 'wb') as f: #二进制写入
                f.write(resp.content)
    time.sleep(5)
```
