# 单个页面下载

```python
import requests
import re
import os
import time
import random

os.chdir(r'C:\Users\CDA')
def validatetitle(title):
    rstr='[/\:*?<>|@\n\r]'
    new_title=re.sub(rstr,'_',title)
    return new_title
url='http://api.vc.bilibili.com/board/v1/ranking/top?page_size=10&next_offset=&tag=%E4%BB%8A%E6%97%A5%E7%83%AD%E9%97%A8&platform=pc'
header={'User-Agent':'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.90 Safari/537.36'}
response = requests.get(url,headers=header)
html=response.json()
#以获得网页源码

items =html['data']['items']
for i in items:
    file_name=i['item']['description']
    file_name=validatetitle(file_name)
    print(file_name)
    video_url=i['item']['video_playurl']
    print(video_url)
    
    resp = requests.get(video_url,headers=header)#获取视频内容
    video_path = '{file_name}.mp4'.format(file_name=video_name)
    with open(video_path, 'wb') as f:
        f.write(resp.content)
    time.sleep(random.randint(5,8))
```

# 多个页面下载

```python
import requests
import re
import os
import time
import random
os.chdir(r'C:\Users\CDA')

def validatetitle(title):  #文件名格式化
    rstr='[/\:*?<>|@\n\r]'
    new_title=re.sub(rstr,'_',title)
    return new_title
header={'User-Agent':'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.90 Safari/537.36'}

start=0
while True:
    try:
        if start==0:
            url='http://api.vc.bilibili.com/board/v1/ranking/top?page_size=10&next_offset={}&tag=%E4%BB%8A%E6%97%A5%E7%83%AD%E9%97%A8&platform=pc'.format('')
        else:
            url='http://api.vc.bilibili.com/board/v1/ranking/top?page_size=10&next_offset={}&tag=%E4%BB%8A%E6%97%A5%E7%83%AD%E9%97%A8&platform=pc'.format(start*10+1)
        print(url)
        response = requests.get(url,headers=header)
        html=response.json()
#获得网页源码

        items =html['data']['items']
        for i in items:
            video_name=i['item']['description']
            video_name=validatetitle(video_name)
            print(video_name)
            video_url=i['item']['video_playurl']
            print(video_url)
#获得片名和片url   

            resp = requests.get(video_url,headers=header)#获取视频内容
            file_path = r'{file_name}.mp4'.format(file_name=video_name)
            with open(file_path, 'wb') as f:
                f.write(resp.content)
            time.sleep(random.randint(2,3))
        start=start+1
    except Exception as e:
        print(e)
        break
```
