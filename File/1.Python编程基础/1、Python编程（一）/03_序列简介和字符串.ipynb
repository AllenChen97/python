# 序列

## 序列的概念

- 序列是一处存放多个值的连续内存空间。
- 序列的每一个值称为元素，每一个元素都会分配一个数字，这个数字成为索引（或位置）。
- 在Python中，序列结构的主要有字符串、列表、元组、字典、集合。（我们后面会陆续介绍）

## 索引

&ensp; 
&ensp;
<font color=#00008B size=3 face="黑体">
**（1）从左往右**

- 序列中每一个元素都有一个编号，称为索引。
- 这个索引是从0开始递增的，下标为0表示第一个元素，下标为2表示第2个元素，如此类推。

举个例子，比如字符串a="我爱北京天安门"，索引如下：

如果我想截取出“爱”字，就可以直接使用“爱”对应的索引值“1”：

````python
a="我爱北京天安门"
a[-2::-1]
````

&ensp;

&ensp;
<font color=#00008B size=3 face="黑体">
**（2）从右往左**

第二种索引是从右到左，即最后一个元素的索引值是-1，倒数第二个是-2，如此类推。


那么，如果还是字符串a="我爱北京天安门"，取出“爱”字就还可以使用另一种索引方法：

````python
a="我爱北京天安门"
a[-6]
````

## 切片

<img width = '500' height ='400' src ="https://timgsa.baidu.com/timg?image&quality=80&size=b9999_10000&sec=1547976524712&di=443267991eb98b36b00c25f88b1a12f5&imgtype=0&src=http%3A%2F%2F5b0988e595225.cdn.sohucs.com%2Fq_70%2Cc_zoom%2Cw_640%2Fimages%2F20180103%2F6be6201a61f3423ebdeadee9ca856f3c.gif"/>

- 如果我们想截取的是“我爱北京天安门”其中的“天安门”两个字  
使用单个索引的方法就无法实现，此时我们可以使用切片操作  
- 切片可以访问一定范围的元素
- 通过切片返回的是一个新的序列

切片语法格式：  
<font color=#DC143C size=4 face="黑体">  
    
**sname[start: end :step]**
<font color=#000000 size=2 face="宋体">
- **sname:序列的名称**
- **start：切片的开始位置，包括该位置（如果不写，默认为0或-1）**
- **end：切片的截止位置，不包括该位置（如果不写，默认是start的另一端）**
- **step：切片步长，不写的话默认为1**,**<font color=#DC143C> （如果是正数意味着从左取到右，负数意味着从右取到左）**

比如，还是a="我爱北京天安门"  
- 如果我们取出“天安门”这三字，start就是“天”（索引值为4）
- end是“门”（索引值为6，但是不包括6，还需要再往前一步取，end的索引值需要是7）
- 因为我们是一个个取，不是“越位”“跳着”取，且从左到右取，因此步长step为1

<font color=#DC143C size=2 face="黑体">
注意，由于“天安门”这三个字是从左到右取，决定了步长一定是正数，且为1，即使我们换一种索引值来取也是一样。

````python
a="我爱北京天安门"

a[::1]
````

**如果我们想要取完所有字符，可以开始位置和截止位置都不写。**

````python
a="我爱北京天安门"
a[::]
````

**如果要取出“我北天门”这四个字（跳一个字符来读取），可以设置步长为2。**

````python
a="我爱北京天安门"
a[::2]
````

**相反的方向取，即'门天北我'，步长2变成-2就可以。**

````python
a="我爱北京天安门"
a[::-2]
````

## 字符串检索

<font color=#00008B size=3 face="黑体"> 
**语法：

<font color=#000000 size=3 face="黑体">

## 使用%操作符
- 格式化字符串时，Python使用一个字符串作为模板。
- 模板中有格式符（如%s和%d），这些格式符为真实值预留位置，并说明真实数值应该呈现的格式。
- Python用一个tuple将多个值传递给模板，每个值对应一个格式符。

````python
print("I'm John. I'm 2*13 year old")
````

如果想将2*3的运算结果输入到字符串里面，怎么办？

````python
print("I'm %s. I'm %d year old" % ('John', 2*13))
````

上面的例子中，
- ('John', 2*13)的两个元素'Vamei'和99为替换%s和%d的真实值。 
- 在模板和元组之间，有一个%号分隔，它代表了格式化操作。

如果我们将%后面括号的元组(就是%后面的括号内容)赋值到一个变量，就会有实用的效果：

````python
p_01=("John",26)
p_02=("Vamei",99)

print("I'm %s. I'm %d year old" % p_01)
print("I'm %s. I'm %d year old" % p_02)
````

&ensp; 

 
&ensp;
<font color=#DC143C size=3 face="黑体">
**但是这有个问题就是，如果要输入的参数太多，输入顺序记错的话，岂不是很容易混淆？**

解决的办法是将上面的元组替换成字典：

````python
p_01={"name":"John","age":26,"city":"guangzhou"}

print("I'm %(name)s. I'm %(age)d year old.I'm from %(city)s." % p_01)
````

换成字典之后，就可以不用注意输入顺序了

````python
p_02={"city":"shenzhen","name":"Vamei","age":99}

print("I'm %(name)s. I'm %(age)d year old,I'm from %(city)s." % p_02)
````

%s,%d分别表示字符串以str（），以及十进制整数表示，%f表示结果为浮点型，其他不常用。

## 使用字符串对象的format()方法

- 字符串的格式化   
    - 将其他类型的数据转换为字符串，通过format()方法实现   
- 三种格式   
    - **按默认从左到右的默认顺序传值：**  
            “字符串-｛｝ -字符串-｛｝” .format(“内容1” ，“内容2”)   
    - **按指定位置传值：**   
            “字符串-｛2｝ -字符串-｛1｝” .format(“内容1” ，“内容2”)   
    - **按设置参数传值，返回格式化后的字符串：**
            “字符串-｛变量名1｝ -字符串-｛变量名2｝” .format(变量名1 = “内容 ”，变量名2 = “内容2”)   

````python
"{}{}{}".format("who ","am ","i ")
````

````python
"{2}{1}{0}".format("who ","am ","i ")
````

````python
"{a}{b}{c}".format(c="who ",b="am ",a="i ")
````
