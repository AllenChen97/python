# 梯度下降法求解最小值

```python
#画图
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x = np.linspace(-5,9)#等差数列
y = x**2-4*x
plt.figure()
plt.plot(x,y)
plt.grid()
plt.title('y=x^2-4x')
```

```python
#梯度下降法求解最小值
#f(x)的导数
def f_prime(x_old):
    return 2 * x_old - 4

x_old = 8#初始值
alpha = 0.001#步长，也就是学习速率，控制更新的幅度
for i in range(10000):
    x_new = x_old - alpha * f_prime(x_old)#上面提到的公式
    x_old = x_new
print(x_new)
```

# 梯度下降法

```python
#梯度下降法
import os
os.chdir('C:\\Users\\CDA')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
# y = kx + b,k is slope, b is y-intercept

def step_gradient(b_current, k_current, points, learningRate):
    b_gradient = 0
    k_gradient = 0
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        k_gradient += (((k_current * x) + b_current)-y)*x
        b_gradient += (((k_current * x) + b_current)-y)*1
    new_b = b_current - (learningRate * b_gradient)
    new_k = k_current - (learningRate * k_gradient)
    return [new_b, new_k]

#误差函数
def compute_error_for_line_given_points(b, k, points):
    totalError = 0
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        totalError += (y - (k * x + b)) ** 2
    return totalError


def gradient_descent_runner(points, starting_b, starting_k, learning_rate, num_iterations):
    b = starting_b
    k = starting_k
    error=[]
    for i in range(num_iterations):
        b, k = step_gradient(b, k, np.array(points), learning_rate)
        error.append(compute_error_for_line_given_points(b, k, points))
    return [b, k,error]
    


#画原始数据图
points = np.array(pd.read_table("data.csv", delimiter=",",names=['x1','y']))
plt.figure(figsize=(8,4),dpi=100)
plt.scatter(points[:,0],points[:,1],color='black')
plt.title("original_data")


#梯度下降法
plt.figure(figsize=(8,4),dpi=100)
plt.scatter(points[:,0],points[:,1],color='black')

learning_rate = 0.000001  #学习率
initial_b = 0 # initial y-intercept guess
initial_k = 0 # initial slope guess
num_iterations = 10000
[b, k, error] = gradient_descent_runner(points, initial_b, initial_k, learning_rate, num_iterations)
y_predict=points[:,0]*k+b
plt.plot(points[:,0],y_predict,color='blue')
plt.title("GradientDescent")

#迭代次数与误差的关系
plt.figure(figsize=(8,4),dpi=100)
plt.plot(range(num_iterations)[0:100],np.array(error)[0:100],color='red')
plt.xlabel("iteration number")                                   #添加x轴的名称
plt.ylabel("error") 
```

```python
#BGD:
import random
#用y = Θ1*x1 + Θ2*x2来拟合下面的输入和输出
input_x = [[1,4],[2,5],[5,1],[4,2],[3,4],[1,1],[0,1],[2,1],[0,0],[1,0]]  
y = [19,26,19,20,26,7,4,11,0,3] 
theta = [1,1]       #θ参数初始化
loss = 10           #loss先定义一个数，为了进入循环迭代，比较损失函数值
step_size = 0.0001  #学习率
eps =1             #误差精度
error =0           #每个点的误差
iter_count = 0     #当前迭代次数
while( loss > eps):
    loss = 0
    #这里更新权重的时候所有的样本点都用上了
    for i in range (10):
        pred_y = theta[0]*input_x[i][0]+theta[1]*input_x[i][1]
        theta[0] = theta[0] - step_size * (pred_y - y[i]) * input_x[i][0]
        theta[1] = theta[1] - step_size * (pred_y - y[i]) * input_x[i][1]
        error = 0.5*(pred_y - y[i])**2
        loss = loss + error
    iter_count += 1

print('theta: ',theta) 
print('final loss: ', loss)
print('iters: ', iter_count*10)


#SGD:
#每次选取一个随机值，随机选择一个点更新θ
import random
#This is a sample to simulate a function y = theta1*x1 + theta2*x2
input_x = [[1,4],[2,5],[5,1],[4,2],[3,4],[1,1],[0,1],[2,1],[0,0],[1,0]]  
y = [19,26,19,20,26,7,4,11,0,3] 
theta = [1,1]       #θ参数初始化
loss = 10           #loss先定义一个数，为了进入循环迭代，比较损失函数值
step_size = 0.0001  #学习率
eps =1             #误差精度
error =0           #每个点的误差
iter_count = 0     #当前迭代次数
while( loss > eps):
    loss = 0
    #每一次选取随机的一个点进行权重的更新
    i = random.randint(0,9)
    pred_y = theta[0]*input_x[i][0]+theta[1]*input_x[i][1]
    theta[0] = theta[0] - step_size * (pred_y - y[i]) * input_x[i][0]
    theta[1] = theta[1] - step_size * (pred_y - y[i]) * input_x[i][1]

    #计算总的误差
    for i in range (10):
        pred_y = theta[0]*input_x[i][0]+theta[1]*input_x[i][1]
        error = 0.5*(pred_y - y[i])**2
        loss = loss + error
    iter_count += 1
    #print('iters_count', iter_count)

print('theta: ',theta) 
print('final loss: ', loss)
print('iters: ', iter_count*1)


#MBGD:
#用了2个样本点
import random
#This is a sample to simulate a function y = theta1*x1 + theta2*x2
input_x = [[1,4],[2,5],[5,1],[4,2],[3,4],[1,1],[0,1],[2,1],[0,0],[1,0]]  
y = [19,26,19,20,26,7,4,11,0,3] 
theta = [1,1]       #θ参数初始化
loss = 10           #loss先定义一个数，为了进入循环迭代，比较损失函数值
step_size = 0.0001  #学习率
eps =1             #误差精度
error =0           #每个点的误差
iter_count = 0     #当前迭代次数
while( loss > eps):
    loss = 0
    i = random.randint(0,9) #注意这里，我这里批量每次选取的是2个样本点做更新，另一个点是随机点+1的相邻点
    j = (i+1)%10
    pred_y = theta[0]*input_x[i][0]+theta[1]*input_x[i][1]
    theta[0] = theta[0] - step_size * (pred_y - y[i]) * input_x[i][0]
    theta[1] = theta[1] - step_size * (pred_y - y[i]) * input_x[i][1]

    pred_y = theta[0]*input_x[j][0]+theta[1]*input_x[j][1]
    theta[0] = theta[0] - step_size * (pred_y - y[j]) * input_x[j][0]
    theta[1] = theta[1] - step_size * (pred_y - y[j]) * input_x[j][1]  
    for i in range (10):
        pred_y = theta[0]*input_x[i][0]+theta[1]*input_x[i][1]
        error = 0.5*(pred_y - y[i])**2
        loss = loss + error
    iter_count += 1
    #print('iters_count', iter_count)

print('theta: ',theta) 
print('final loss: ', loss)
print('iters: ', iter_count*2)
```

# 全局牛顿法（了解）

```python
#全局牛顿法
import numpy as np
import pandas as pd
from math import pow
import matplotlib.pyplot as plt
def load_data():
    data=pd.read_table("data_line.txt",sep='	',names=['x1','y'])
    data['x0']=1
    feature=data[['x0','x1']]
    label=data['y']
    return np.mat(feature), np.mat(label).T

def least_square(feature, label):
    '''最小二乘法
    input:  feature(mat):特征
            label(mat):标签
    output: w(mat):回归系数
    '''
    w = (feature.T * feature).I * feature.T * label
    return w

def first_derivativ(feature, label, w):
    '''计算一阶导函数的值
    input:  feature(mat):特征
            label(mat):标签
    output: g(mat):一阶导数值
    '''
    m, n = np.shape(feature)
    g = np.mat(np.zeros((n, 1)))
    for i in range(m):
        err = label[i, 0] - feature[i, ] * w
        for j in range(n):
            g[j, ] -= err * feature[i, j]
    return g     

def second_derivative(feature):
    '''计算二阶导函数的值
    input:  feature(mat):特征
    output: G(mat):二阶导数值
    '''
    m, n = np.shape(feature)
    G = np.mat(np.zeros((n, n)))
    for i in range(m):
        x_left = feature[i, ].T
        x_right = feature[i, ]
        G += x_left * x_right
    return G

def get_error(feature, label, w):
    '''计算误差
    input:  feature(mat):特征
            label(mat):标签
            w(mat):线性回归模型的参数
    output: 损失函数值
    '''
    return (label - feature * w).T * (label - feature * w) / 2

def get_min_m(feature, label, sigma, delta, d, w, g):
    '''计算步长中最小的值m
    input:  feature(mat):特征
            label(mat):标签
            sigma(float),delta(float):全局牛顿法的参数
            d(mat):负的一阶导数除以二阶导数值
            g(mat):一阶导数值
    output: m(int):最小m值
    '''
    m = 0
    while True:
        w_new = w + pow(sigma, m) * d
        left = get_error(feature, label , w_new)
        right = get_error(feature, label , w) + delta * pow(sigma, m) * g.T * d
        if left <= right:
            break
        else:
            m += 1
    return m           

def newton(feature, label, iterMax, sigma, delta):
    '''牛顿法
    input:  feature(mat):特征
            label(mat):标签
            iterMax(int):最大迭代次数
            sigma(float), delta(float):牛顿法中的参数
    output: w(mat):回归系数
    '''
    n = np.shape(feature)[1]
    w = np.mat(np.zeros((n, 1)))
    it = 0
    while it <= iterMax:
        # print it
        g = first_derivativ(feature, label, w)  # 一阶导数
        G = second_derivative(feature)  # 二阶导数
        d = -G.I * g
        m = get_min_m(feature, label, sigma, delta, d, w, g)  # 得到最小的m
        w = w + pow(sigma, m) * d
        it += 1       
    return w

if __name__ == "__main__":
    # 导入数据集
    feature, label = load_data()
    feature1=feature[:,1].getA()
    label1=label.getA()
    plt.figure(figsize=(8,4),dpi=100)
    plt.scatter(feature1,label1,color='black')
    plt.title("original_data")
    
    # 最小二乘求解
    w_ls = least_square(feature, label)
    y_fit1=feature * w_ls
    plt.figure(figsize=(8,4),dpi=100)
    plt.scatter(feature1,label1,color='black')
    plt.plot(feature1,y_fit1,color ='blue',linewidth=3)
    plt.title("least_square")
    
    # 牛顿法求解
    w_newton = newton(feature, label, 50, 0.1, 0.5)
    plt.figure(figsize=(8,4),dpi=100)
    y_fit2=feature * w_newton
    plt.scatter(feature1,label1,color='black')
    plt.plot(feature1,y_fit2,color ='red',linewidth=3)  
    plt.title("newton")
```

```python
#岭回归
%matplotlib inline
import os
os.chdir('C:\\Users\\CDA\\data\\线')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
def load_data():
    data=pd.read_table('data_ridge.txt',sep='	',names=['x1','y'])
    data['x0']=1
    feature=data[['x0','x1']]
    label=data['y']
    return np.mat(feature), np.mat(label).T

def ridge_regression(feature, label, lam):
    '''最小二乘法求解
    input:  feature(mat):特征
            label(mat):标签
    output: w(mat):回归系数
    '''
    n = np.shape(feature)[1]
    w = (feature.T * feature + lam * np.mat(np.eye(n))).I * feature.T * label
    return w


if __name__ == "__main__":
    # 导入数据
    feature, label = load_data()
    plt.figure(figsize=(8,4),dpi=100)
    feature1=feature[:,1].getA()
    label1=label.getA()
    plt.scatter(feature1,label1,color='black')
    
    # 训练模型
    w0 = ridge_regression(feature, label, 0.001)#lam参数,可以调整
    y_fit1=feature * w0
    plt.plot(feature1,y_fit1,color ='blue',linewidth=3)
    plt.title("ridge_regression")
```

# 岭回归曲线绘制

```python
#岭回归系数选择
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from math import exp
import os
os.chdir('C:\\Users\\chen\\File')
def load_data():
    data=pd.read_table('abalone.txt',sep='	',names=['x1','x2','x3','x4','x5','x6','x7','x8','y'])
    feature=data.loc[:,'x1':'x8']
    label=data['y']
    return np.mat(feature), np.mat(label).T

def standarize(X):
    ''' 标准化数据 (零均值, 单位标准差)
    '''
    std_deviation = np.std(X)
    mean = np.mean(X)
    return (X - mean)/std_deviation

def ridge_regression(X, y, lambd=0.2):
    ''' 获取岭回归系数
    '''
    m,n = X.shape
    I = np.matrix(np.eye(n))
    w = (X.T*X + lambd*I).I*X.T*y
    return w

def ridge_traj(X, y, k=20):
    ''' 获取岭轨迹矩阵
    '''
    m, n = X.shape
    ws = np.zeros((k, n))
    for i in range(k):
        w = ridge_regression(X, y, lambd=exp(i-10))
        ws[i, :] = w.T
    return ws

if '__main__' == __name__:
    # 加载数据
    X, y = load_data()

    # 标准化
    X, y = standarize(X), standarize(y)
    # 绘制岭轨迹
    k = 30
    ws = ridge_traj(X, y, k)
    lambdas = [exp(i-10) for i in range(k)]
    plt.semilogx(lambdas, ws)
    plt.show()
```

# lasso回归

```python
#lasso回归
import itertools
from math import exp
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
os.chdir('C:\\Users\\chen\\File')
def load_data():
    ''' 加载数据
    '''
    data=pd.read_table('abalone.txt',sep='	',names=['x1','x2','x3','x4','x5','x6','x7','x8','y'])
    feature=data.loc[:,'x1':'x8']
    label=data['y']
    return np.mat(feature), np.mat(label).T

def standarize(X):
    ''' 标准化数据 (零均值, 单位标准差)
    '''
    std_deviation = np.std(X)
    mean = np.mean(X)
    return (X - mean)/std_deviation

def lasso_regression(X, y, lambd=0.2, threshold=0.1):
    ''' 通过坐标下降(coordinate descent)法获取LASSO回归系数
    '''
    # 计算残差平方和
    rss = lambda X, y, w: (y - X*w).T*(y - X*w)

    # 初始化回归系数w.
    m, n = X.shape
    w = np.matrix(np.zeros((n, 1)))
    r = rss(X, y, w)

    # 使用坐标下降法优化回归系数w
    niter = itertools.count(1)

    for it in niter:
        for k in range(n):
            # 计算常量值z_k和p_k
            z_k = (X[:, k].T*X[:, k])[0, 0]
            p_k = 0
            for i in range(m):
                p_k += X[i, k]*(y[i, 0] - sum([X[i, j]*w[j, 0] for j in range(n) if j != k]))

            if p_k < -lambd/2:
                w_k = (p_k + lambd/2)/z_k
            elif p_k > lambd/2:
                w_k = (p_k - lambd/2)/z_k
            else:
                w_k = 0

            w[k, 0] = w_k

        r_prime = rss(X, y, w)
        delta = abs(r_prime - r)[0, 0]
        r = r_prime
        #print('Iteration: {}, delta = {}'.format(it, delta))

        if delta < threshold:
            break

    return w

def lasso_traj(X, y, k=20):
    ''' 获取回归系数轨迹矩阵
    '''
    m, n = X.shape
    ws = np.zeros((k, n))
    for i in range(k):
        w = lasso_regression(X, y, lambd=exp(i-10))
        ws[i, :] = w.T
        #print('lambda = e^({}), w = {}'.format(i-10, w.T[0, :]))
    return ws

if '__main__' == __name__:
    X, y = load_data()
    X, y = standarize(X), standarize(y)
    w = lasso_regression(X, y, lambd=10)
    # 绘制轨迹
    k = 20
    ws = lasso_traj(X, y, k)
    fig = plt.figure()
    lambdas = [exp(i-10) for i in range(k)]
    plt.semilogx(lambdas, ws)
    plt.show()
```

# 案例：输出电力预测

```python
import os
os.chdir('C:\\Users\\CDA\\data\\线')
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model

data = pd.read_csv('ccpp.csv')
data.head()#查看前五行数据，如果是最后五行，用data.tail()
data.tail()
data.shape#看看数据的维度
X = data[['AT', 'V', 'AP', 'RH']]#AT，V，AP和RH这4个列作为样本特征
X.head()
y = data[['PE']]#样本输出y， 我们用PE作为样本输出。
y.head()

#把X和y的样本组合划分成两部分，一部分是训练集，一部分是测试集：
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.1,random_state=1)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)#75%的样本数据被作为训练集，25%的样本被作为测试集

from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train, y_train)#训练
print(linreg.intercept_) #查看截距 theta0
print(linreg.coef_)#求得的其他四个值个值。也就是说PE和其他4个变量的关系如下：
```

```python
#模型评价 MSE或者RMSE，如果我们用其他方法得到了不同的系数
#均方差（Mean Squared Error, MSE）或者均方根差(Root Mean Squared Error, RMSE)
y_pred = linreg.predict(X_test)
from sklearn import metrics
# 用scikit-learn计算MSE
print("MSE:",metrics.mean_squared_error(y_test, y_pred))
# 用scikit-learn计算RMSE
print("RMSE:",np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
#用score评价
regr=linreg.fit(X_train, y_train)#训练
regr.score(X_test, y_test)
```

```python
#采用10折交叉验证，即cross_val_predict中的cv参数为10：
X = data[['AT', 'V', 'AP', 'RH']]
y = data[['PE']]
from sklearn.model_selection import cross_val_predict
predicted = cross_val_predict(linreg, X, y, cv=10)
# 用scikit-learn计算MSE
print("MSE:",metrics.mean_squared_error(y, predicted))
# 用scikit-learn计算RMSE
print("RMSE:",np.sqrt(metrics.mean_squared_error(y, predicted)))
```

```python
#展示
plt.figure(figsize=(8,4),dpi=100)
plt.scatter(y, predicted)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r', lw=2)
plt.xlabel('Measured')
plt.ylabel('Predicted')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV ,train_test_split
from sklearn.datasets import load_boston
from sklearn.metrics import mean_squared_error
ld= load_boston()
x_train,x_test,y_train,y_test = train_test_split(ld.data,ld.target,test_size=0.2)

std = StandardScaler()
x_train = std.fit_transform(x_train)
x_test = std.transform(x_test)
y_train = std.fit_transform(y_train.reshape(-1,1))
y_test = std.transform(y_test.reshape(-1,1))

lr = LinearRegression()
rd = Ridge()
ls = Lasso()
en = ElasticNet()
params1 = {'alpha':[6,7,8,9,10]}
params2 = {'alpha':[0.001,0.003,0.005]}
params3 = {'alpha':[0.008,0.01,0.012,0.014],'l1_ratio':[0.001,0.01,0.05,0.06,0.07]}
gsrd = GridSearchCV(rd,params1,cv=10)
gsls = GridSearchCV(ls,params2,cv=10)
gsen = GridSearchCV(en,params3,cv=10)

lr.fit(x_train,y_train)
print("线性回归评分",lr.score(x_test,y_test))

print("岭回归评分",gsrd.fit(x_train,y_train).score(x_test,y_test))
print(gsrd.best_params_)

print("Lasso回归评分",gsls.fit(x_train,y_train).score(x_test,y_test))
print(gsls.best_params_)

print("弹性网回归评分",gsen.fit(x_train,y_train).score(x_test,y_test))
print(gsen.best_params_)
```

```python
predict1 = LinearRegression().fit(x_train,y_train).predict(x_test)
print("线性回归均方差",mean_squared_error(std.inverse_transform(y_test),predict1))

predict2 = Ridge(alpha=10).fit(x_train,y_train).predict(x_test)
print("岭回归均方差",mean_squared_error(std.inverse_transform(y_test),predict2))

predict3 = Lasso(alpha=0.003).fit(x_train,y_train).predict(x_test)
print("Lasso回归均方差",mean_squared_error(std.inverse_transform(y_test),predict3))

predict4 = ElasticNet(alpha=0.016,l1_ratio=0.01).fit(x_train,y_train).predict(x_test)
print("弹性网回归均方差",mean_squared_error(std.inverse_transform(y_test),predict4))
```

```python
# 2、标准化处理
# 特征值处理
std_x = StandardScaler()
x_train = std_x.fit_transform(x_train)
x_test = std_x.transform(x_test)
# 目标值进行处理
std_y  = StandardScaler()
y_train = std_y.fit_transform(y_train.reshape(-1,1))
y_test = std_y.transform(y_test.reshape(-1,1))
```

```python
# 3、算法
# LinearRegression
lr=LinearRegression().fit(x_train,y_train)
y_lr_predict = lr.predict(x_test)
y_lr_predict = std_y.inverse_transform(y_lr_predict)

from sklearn.metrics import mean_squared_error

print("lr的均方误差为：",mean_squared_error(std_y.inverse_transform(y_test),y_lr_predict))
plt.figure(figsize=(8,4),dpi=100)
plt.title("LinearRegression")
plt.plot(y_lr_predict, std_y.inverse_transform(y_test), 'rx')
plt.plot([y_lr_predict.min(), y_lr_predict.max()], [y_lr_predict.min(), y_lr_predict.max()], 'b-.', lw=4)

plt.ylabel("Predieted Price")
plt.xlabel("Real Price")
plt.show()
```

```python
# 带有正则化的岭回归
rd = Ridge(alpha=0.02)
rd.fit(x_train,y_train)
y_rd_predict = rd.predict(x_test)
y_rd_predict = std_y.inverse_transform(y_rd_predict)
print("Ridge的均方误差为：",mean_squared_error(std_y.inverse_transform(y_test),y_rd_predict))
plt.figure(figsize=(8,4),dpi=100)
plt.title("RidgeRegressor")
plt.plot(y_rd_predict, std_y.inverse_transform(y_test), 'rx') # y = ωX
plt.plot([y_rd_predict.min(), y_rd_predict.max()], [y_rd_predict.min(), y_rd_predict.max()], 'b-.', lw=4) # f(x)=x
plt.ylabel("Predieted Price")
plt.xlabel("Real Price")
plt.show()
```

```python
# 带有正则化的LASSO回归
rd = Lasso(alpha=0.02)
rd.fit(x_train,y_train)
y_rd_predict = rd.predict(x_test)
y_rd_predict = std_y.inverse_transform(y_rd_predict)
print("Ridge的均方误差为：",mean_squared_error(std_y.inverse_transform(y_test),y_rd_predict))
plt.figure(figsize=(8,4),dpi=100)
plt.title("LassoRegressor")
plt.plot(y_rd_predict, std_y.inverse_transform(y_test), 'rx') # y = ωX
plt.plot([y_rd_predict.min(), y_rd_predict.max()], [y_rd_predict.min(), y_rd_predict.max()], 'b-.', lw=4) # f(x)=x
plt.ylabel("Predieted Price")
plt.xlabel("Real Price")
plt.show()
```

# Pipeline: 链式评估器

```python
#Pipeline 使用一系列 (key, value) 键值对来构建,其中 key 是步骤起的名字， value 是一个评估器对象
#https://blog.csdn.net/fontthrone/article/details/78234198
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression, Ridge,Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_boston
from sklearn.pipeline import Pipeline

from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt 
ld = load_boston()
x_train,x_test,y_train,y_test = train_test_split(ld.data,ld.target,test_size=0.25)

estimators = [('sc', StandardScaler()),
                    ('pca', PCA(n_components=10)),
                    ('clf', Ridge(alpha=0.02))
                    ]

pipe_lr = Pipeline(estimators)
pipe_lr.fit(x_train, y_train)
print('Test accuracy: %.3f' % pipe_lr.score(x_test, y_test))
```
