# 数据预处理

```python
Na_list = Na.index.tolist()
Na_vars=[]
for i in Na_list:
    if insur[i].isnull().sum()/len(insur) > 0.75:
        del insur[i]
        
    elif insur[i].dtypes != object :
        insur[i].fillna(insur[i].quantile(0.5),inplace=True)
        
    elif insur[i].dtypes == object and len(insur[i].unique())<=3 :
        insur[i].fillna("N",inplace=True)
    else:
        Na_vars.append(i)
Na_vars
```

```python
insur[Na_vars]
```

```python
insur.drop_duplicates()
```

```python
insur.drop(columns=["KBM_INDV_ID","STATE_NAME"],inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
vars2=[]
for i in insur.columns:
    if len(insur[i].unique()) <3:
        vars2.append(i)

for j in vars2:
    le = LabelEncoder()
    insur[j] = le.fit_transform(insur[j])
insur[vars2].head()
```

```python
for i in Na_vars:
    aa = pd.get_dummies(insur[i],prefix=i)
    for j in aa:
        if j in '_nan':
            del aa[j]
    del insur[i]
    insur = pd.concat([insur,aa],axis=1)
```
