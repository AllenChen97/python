# 鸢尾花数据集降维

```python
#导⼊入模块和包
from sklearn.datasets import load_iris #导⼊入sklearn中内置的鸢尾花数据集
from sklearn.decomposition import PCA #导⼊入PCA降维算法
import matplotlib.pyplot as plt #绘图包
#提取数据集
iris = load_iris()
y = iris.target
X = iris.data
#调⽤用PCA建模
pca = PCA(n_components=2) #实例例化
pca = pca.fit(X) #拟合模型
X_dr = pca.transform(X) #获取新矩阵
#绘制可视化图形
#要将三种鸢尾花的数据分布显示在⼆二维平⾯面坐标系中，对应的两个坐标（两个特征向量量）应该是三种鸢尾
#花降维后的x1和x2，怎样才能取出三种鸢尾花下不不同的x1和x2呢？
X_dr[y == 0, 0] #这⾥里里是布尔索引，看出来了了么？
```

```python
plt.figure()
plt.scatter(X_dr[y==0, 0], X_dr[y==0, 1], c="red", label=iris.target_names[0])
plt.scatter(X_dr[y==1, 0], X_dr[y==1, 1], c="black",
label=iris.target_names[1])
plt.scatter(X_dr[y==2, 0], X_dr[y==2, 1], c="orange",
label=iris.target_names[2])
plt.legend()
plt.title('PCA of IRIS dataset')
plt.show()
```

```python
colors = ['red', 'black', 'orange']
iris.target_names
plt.figure()
for i in [0, 1, 2]:
    plt.scatter(X_dr[y == i, 0]
    ,X_dr[y == i, 1]
    ,alpha=.7
    ,c=colors[i]
    ,label=iris.target_names[i]
    )
plt.legend()
plt.title('PCA of IRIS dataset')
plt.show()
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

# 累计可解释⽅方差贡献率曲线

```python
import numpy as np
pca_line = PCA().fit(X)
plt.plot([1,2,3,4],np.cumsum(pca_line.explained_variance_ratio_))
plt.xticks([1,2,3,4]) #这是为了了限制坐标轴显示为整数
plt.xlabel("number of components after dimension reduction")
plt.ylabel("cumulative explained variance ratio")
plt.show()
```

# 最⼤大似然估计⾃自选超参数

```python
pca_mle = PCA(n_components="mle")
pca_mle = pca_mle.fit(X)
X_mle = pca_mle.transform(X)
X_mle
```

```python
pca_mle.explained_variance_ratio_.sum()
```

# 按信息量量占⽐比选超参数

```python
pca_f = PCA(n_components=0.97,svd_solver="full")
pca_f = pca_f.fit(X)
X_f = pca_f.transform(X)
pca_f.explained_variance_ratio_
```
