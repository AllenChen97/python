# association_rules

```python
#association_rules  http://rasbt.github.io/mlxtend/user_guide/frequent_patterns/association_rules/
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori

dataset = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]

te = TransactionEncoder()
te_ary = te.fit_transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
```

```python
help(association_rules)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2)
rules
```

## 选择与过滤
1.at least 2 antecedents
2.a confidence > 0.75
3.a lift score > 1.2

```python
rules["antecedent_len"] = rules["antecedents"].apply(lambda x: len(x))#前件的长度
rules
```

```python
rules[ (rules['antecedent_len'] >= 2) &
       (rules['confidence'] > 0.75) &
       (rules['lift'] > 1.2) ]
```

# 案例分析，英国零售商数据分析¶
此处的具体数据来自UCI机器学习库，代表了2010-2011英国零售商的交易数据。  
这主要代表对批发商的销售，因此与消费者购买模式略有不同，但仍然是一个有用的案例研究。

# 这个数据集中有8个属性，分别为：
InvoiceNo：发票号码，作为交易记录的唯一标识；
StockCode：商品代码；
Description：商品描述；
Quantity：交易数量；
InvoiceDate：开票日期，作为交易产生的日期；
UnitPrice：商品单价；
CustomerID：客户ID；
Country：客户的国家；

```python
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
import warnings
warnings.filterwarnings("ignore")
import os
os.chdir('C:\\Users\\CDA\\2.推荐系统')
```

```python
#df = pd.read_excel('http://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx')
df = pd.read_excel('Online Retail.xlsx')
print('The data size is:', df.shape)
df.head()
```

```python
df.info()
```

```python
df['Description'][1:2].values
```

数据清洗。首先，需要删除一些描述包含的空格，删除没有发票编号的行，以及移除那些信贷交易(those with invoice numbers containing C).

```python
df['Description'] = df['Description'].str.strip()
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)#原地删除
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
```

```python
df['InvoiceNo'].str.contains('C')
```

```python
df = df[~df['InvoiceNo'].str.contains('C')]
df.head()
```

清理之后，将每个项目合并为每行1个事务，每个产品进行one-hot编码。
为了获得小数据集，只选用了法国的销售情况。

```python
#unstack:https://www.cnblogs.com/bambipai/p/7658311.html
```

```python
df[df['Country'] == "France"].groupby(['InvoiceNo', 'Description'])['Quantity'].sum()
```

```python
df[df['Country'] == "France"].groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().reset_index()
```

```python
basket = (df[df['Country'] == "France"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
basket.head()
```

保证>0的值转换为1，<=0的值转换为0。从而保证one-hot编码。  
删除“POSTAGE”列（因为这笔费用不是我们想要探索的费用）

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1


basket_sets = basket.applymap(encode_units)
basket_sets.drop('POSTAGE', inplace=True, axis=1)

basket.head()
```

使用算法进行关联运算

```python
from mlxtend.frequent_patterns import apriori, association_rules
#最小支持度 找出频繁集
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
#最小提升度 找出关联规则
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=3)

rules.sort_values("lift",ascending=False).head()
```

```python
# 我们可以分别对提升度Lift和置信度Confidence进行筛选
selected_rules = rules[ (rules['lift'] >= 6) & (rules['confidence'] >= 0.8) ]
selected_rules
```

```python
support = rules.as_matrix(columns=['support'])
confidence = rules.as_matrix(columns=['confidence'])
plt.title('Association Rules')
plt.xlabel('support')
plt.ylabel('confidence')
plt.scatter(support, confidence)
```
