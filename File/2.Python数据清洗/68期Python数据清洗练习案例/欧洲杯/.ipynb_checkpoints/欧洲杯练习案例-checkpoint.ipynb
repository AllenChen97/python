````python
##  综合题
- euro12中有哪些字段属于object类?
- 提取带百分号字符的字段到一个新的DataFrame中
- 在新DataFrame中，将带有百分号的字符转化为浮点型数据
- 在上表所有列标签中后面备注——（%）
````

````python
eu.describe(include=["object"])
````

````python
title=eu.describe(include=["object"]).iloc[:,1:].columns.tolist()
eu1=eu[title]
eu1
````

````python
eu1
def tran(x):
    return float(x[:-1])/100

eu1.applymap(tran)
````

````python
title[1] = "Goals-to-shots"
eu1.columns=title
````

````python
def add(x):
    return x + r"(%)"

eu1.columns=eu1.columns.map(add)
eu1
````
